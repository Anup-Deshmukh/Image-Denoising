## Load modules

```python
import matplotlib.pyplot as plt

from skimage.color import rgb2gray
from skimage import data
from skimage.transform import resize

# Numpy is useful for handling arrays and dense matrices (a matrix with a lot of nonzeros).
import numpy as np
```

## Load image

```python
img = data.stereo_motorcycle()[0]
img = rgb2gray(img)*255 # convert to gray and change scale from (0,1) to (0,255).

m = img.shape[0]
n = img.shape[1]

#print(m)
#print(n)
plt.figure(1, figsize=(10, 10))
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
#plt.show()
```

## Question 1: compute the differences operators here. Note that the image is not a square image. Marks: 10/100

```python
# You will need these three methods to construct the sparse differences operators.
# If you do not use sparse operators you might have memory problems.
from scipy import sparse
from scipy.sparse import diags
from scipy.sparse import kron
from scipy.sparse import identity


# Start coding here.
#vectorize the image
vec_img = img.flatten('F')
#print(vec_img.shape)

J = -identity(n) + diags([1] * (n-1), 1)

I = identity(m)

D_h = kron(J, I)

D_v = kron(I, J)

#HFD = sparse.csr_matrix(D_h).multiply(vec_img).tocsr()
HFD = sparse.csr_matrix(D_h).dot(vec_img)
print(sparse.csr_matrix(HFD).shape)

#VFD = sparse.csr_matrix(D_v).multiply(sparse.csr_matrix(vec_img)).tocsr()
VFD = sparse.csr_matrix(D_v).dot(vec_img)
print(sparse.csr_matrix(VFD).shape)

```

## Question 2: visualize the differences operators applied on the image. Marks: 5/100

```python
#Check the figures in the lecture notes on piazza to see an example of what the output should be like. 

# Start coding here.

HFD_plot = sparse.csr_matrix(HFD).toarray()
HFD_plot = HFD_plot.flatten('F')
HFD_plot = np.reshape(HFD_plot, (500, 741), order='F')

print("HFD plot")
plt.figure(1, figsize=(10, 10))
plt.imshow(HFD_plot, cmap='gray', vmin=0, vmax=255)
plt.show()

#####
print("VFD plot")
VFD_plot = sparse.csr_matrix(VFD).toarray()
VFD_plot = VFD_plot.flatten('F')
VFD_plot = np.reshape(VFD_plot, (500, 741), order='F')

plt.figure(1, figsize=(10, 10))
plt.imshow(VFD_plot, cmap='gray', vmin=0, vmax=255)
plt.show()
```

## Add noise to the image

```python
mean_ = 0
standard_deviation = 30
dimensions = (m,n)

noise = np.random.normal(mean_,standard_deviation,dimensions)

noisy_image = img + noise

plt.figure(1, figsize=(10, 10))
plt.imshow(noisy_image, cmap='gray', vmin=0, vmax=255)
plt.show()
```

## Quetion 3: Denoise the image. Marks: 20/100

```python

# This method can be used to solve linear systems.
from scipy.sparse.linalg import spsolve
# This module can be used to convert arrays and matrices from complex to real.
from scipy import real

# Use the following lambda list to tune the denoising problem.
lambda_list = [0.1,3,10]


# Start coding here. Check figures in Lecture to see an example of what the output should be like.

for l in lambda_list:
    print("Denoised image with lambda =")
    print(l)
    identity_ls = identity(m*n)
    D_h_tran = D_h.transpose()
    #print(D_h_tran.shape)
    D_v_tran = D_v.transpose()

    term1 = np.dot(D_h_tran, D_h)
    term2 = np.dot(D_v_tran, D_v)
    Dconj_D = l*(term1 + term2)
    #print(sparse.csr_matrix(Dconj_D).shape)

    A_ls = Dconj_D + identity_ls
    #print(sparse.csr_matrix(A_ls).shape)

    vec_noisy_image = noisy_image.flatten('F')
    B_ls = sparse.csr_matrix(vec_noisy_image).transpose()
    #print(sparse.csr_matrix(B_ls).shape)

    denoised_image = sparse.linalg.spsolve(A_ls, B_ls)
    denoised_image = sparse.csr_matrix(denoised_image).transpose()
    #print(sparse.csr_matrix(denoised_image).shape)

    final_image = sparse.csr_matrix(denoised_image).toarray()
    final_image = final_image.flatten('F')
    #print(final_image.shape)
    final_image = np.reshape(final_image, (m, n), order='F')
    #print(final_image.shape)

    #final_image = rgb2gray(final_image)*255
    plt.figure(1, figsize=(10, 10))
    plt.imshow(final_image, cmap='gray', vmin=0, vmax=255)
    plt.show()
```
